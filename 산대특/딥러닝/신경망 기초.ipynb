```python
초기동작 벡터 w = (팔을 돌리는 속도, 각도) 초기화
while(true)
    w에 따라서 수영을 하고 w의 점수 j(w)를 계산
    if(j(w) 원하는 값) break
    더 나은 방향으로 w+ 를 계산
    w = w + w+
w save    
```

```python
# 신경망의 학습은 최적화 문제 --> 최적화의 대상은 가중치(매개변수)
```

```python
입력:훈련데이터
출력 : 최적의 매개변수 값

난수로 매개변수 벡터 W를 초기화 (w는 신경망의 가중치  w0 w1 w2 ....)    
while(true):
    w에 따라 데이터를 인식하고 손실함수 j(w)를 계산
    if j(w) is best  break
    j(w)값을 낮추는 방향으로 w+ 계산
    w = w + w+   --> 가중치를 업데이트
w save    
```

```python
from sklearn.model_selection import validation_curve
import time
```

```python
digit = datasets.load_digits()
x_train,x_test,y_train,y_test = train_test_split(digit.data, digit.target, random_state=32)
```

```python
x_train.shape, x_test.shape
```

```python
# 시작 종료 시간측정
start = time.time()
mlp = MLPClassifier(learning_rate_init=0.001,batch_size=32,max_iter=300,solver='sgd')
prange = range(50,1001,50)
train_score, test_score =  validation_curve(mlp,x_train,y_train,param_name='hidden_layer_sizes',param_range=prange,cv=10,scoring="accuracy",n_jobs=6)
end = time.time()
print(f"하이퍼 매개변수 최적화에 걸린시간 {end-start}초")
```

```python
# 교차검증 결과의 평균화 분산
train_mean = np.mean(train_score, axis=1)
train_std = np.std(train_score, axis=1)

test_mean = np.mean(test_score, axis=1)
test_std = np.std(test_score, axis=1)
```

```python
# 성능 그래프 
import matplotlib.pyplot as plt
plt.plot(prange, train_mean,label='Train score', color='r')
plt.plot(prange, test_mean,label='Test score', color='b')
plt.fill_between(prange,train_mean-train_std, train_mean+train_std, alpha=0.2, color='r') 
plt.fill_between(prange,test_mean-test_std, test_mean+test_std, alpha=0.2, color='b') 
plt.legend()
plt.grid(axis='both')
plt.show()
```

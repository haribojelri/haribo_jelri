# 군집분석

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
```

```python
x = df2.iloc[:,:-1]
y = df2.iloc[:,-1]
rb = RobustScaler()
x_tr = rb.fit_transform(x)
# y = tf.keras.utils.to_categorical(y,2)
```

```python
km =  KMeans(n_clusters=2, random_state=32)
km.fit(x_tr)
```

```python
predict =  km.predict(x_tr)
```

```python
predict.shape, y.shape
```

```python
confusion_matrix = np.zeros((2,2))
confusion_matrix
```

```python
np.unique(predict), np.unique(y)
```

```python
yy = y.values
for i in range(len(predict)):
    confusion_matrix[ predict[i] ][ yy[i] ] += 1
```

```python
total = confusion_matrix[0,0] + confusion_matrix[1,1]
print(f"accuracy = {total / len(predict)}")
```

```python
# 신경망을 개발해서 잘 맞추는 지 확인  분류 문제로  손실함수는 binary_crossentropy
# one hot y를 사용
# L  = 5  인 신경망
# 하이퍼 파라메터
n_input = 11
n_hidden1 = 1024
n_hidden2 = 512
n_hidden3 = 512
n_hidden4 = 512
n_output = 1

from sklearn.model_selection import  train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_tr,y,random_state=32)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(n_hidden1, activation='relu', input_shape = (n_input,)))
model.add(tf.keras.layers.Dense(n_hidden2, activation='relu'))
model.add(tf.keras.layers.Dense(n_hidden3, activation='relu'))
model.add(tf.keras.layers.Dense(n_hidden4, activation='relu'))
model.add(tf.keras.layers.Dense(n_output, activation='sigmoid'))
```

```python
callbacks = [
    tf.keras.callbacks.EarlyStopping(patience=10),
    tf.keras.callbacks.ModelCheckpoint('cpr.keras',save_best_only=True)
]
model.compile(loss = tf.keras.losses.binary_crossentropy, optimizer='adam', metrics = ['acc'])
hist = model.fit(x_train,y_train,batch_size=32, epochs = 100, validation_data = (x_test,y_test),verbose = 2, callbacks=callbacks)
```

```python
model = tf.keras.models.load_model('cpr.keras')
model.evaluate(x_train,y_train),model.evaluate(x_test,y_test)
```

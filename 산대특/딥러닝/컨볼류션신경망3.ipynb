```python
#영상을 증대하기
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
```

```python
# cifar-10 부류
class_name = ['airplane','automobile','bird','cat','deer','dog','flog','horse','ship','truck']
```

```python
# 데이터 취득
(x_train,y_train),(x_test,y_test) = cifar10.load_data()
x_test = x_test.astype('float32') / 255.0
# 테스트 데이터 중에 앞에 12개만 가지고 generator
```

```python
x_test12,y_test12 = x_test[:12], y_test[:12]
```

```python
fig,ax = plt.subplots(1,12,figsize=(16,2))
for i in range(12):
  ax[i].imshow(x_test12[i] )
  ax[i].axis('off')
  ax[i].set_title(class_name[int(y_test[i])] )
```

```python
#영상 증대기 실행
batch_size = 12  # 한번에 생성하는 양
generator = ImageDataGenerator(rotation_range=30.0,width_shift_range=0.2,height_shift_range=0.2,horizontal_flip=True)
```

```python
gen = generator.flow(x_test12,y_test12,batch_size=batch_size)
```

```python
# 1번 generate
img,label =  gen.next()  # next함수는 호출될때마다 Batch Size 매개변수가 지정한 수 만큼 영상 생성
fig, ax = plt.subplots(2,6,figsize=(16,2))
for i in range(batch_size):
  ax[i//6, i % 6].imshow(img[i],cmap='gray_r')
  ax[i//6, i % 6].axis('off')
  ax[i//6, i % 6].set_title(class_name[int(label[i])])
plt.show()                            
```

```python
# 2번 generate
img,label =  gen.next()  # next함수는 호출될때마다 Batch Size 매개변수가 지정한 수 만큼 영상 생성
fig, ax = plt.subplots(2,6,figsize=(16,2))
for i in range(batch_size):
  ax[i//6, i % 6].imshow(img[i],cmap='gray_r')
  ax[i//6, i % 6].axis('off')
  ax[i//6, i % 6].set_title(class_name[int(label[i])])
plt.show()  
```

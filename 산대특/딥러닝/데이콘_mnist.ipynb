```python
import pandas as pd
```

```python
df = pd.read_csv('./mnist/train.csv',index_col=0)
x = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
x.shape, y.shape
```

```python
import numpy as np
np.sqrt(784)
```

```python
from sklearn.linear_model import Perceptron   # 단층 퍼셉트론
from sklearn.neural_network import MLPClassifier  # 다층 퍼셉트론
```

```python
# 데이터 전처리
x = x.astype('float32') / 255.
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=32)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=500, learning_rate_init=0.001,batch_size=32,max_iter=300,solver='sgd',verbose=True)
mlp.fit(x_train,y_train)
mlp.score(x_test,y_test)
```

```python
mlp.score(x_train,y_train), mlp.score(x_test, y_test)
```

```python
test = pd.read_csv('./mnist/test.csv',index_col=0)
```

```python
mlp.predict(test)
```

```python
submission = pd.read_csv('./mnist/sample_submission.csv',index_col=0)
```

```python
submission['label'] = mlp.predict(test)
```

```python
submission.to_csv('submission.csv')
```

```python
# 텐서
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,MaxPooling2D,Conv2D,Flatten
```

```python
C C  P D -  C C  P D
model = Sequential()
model.add(Conv2D(256,(3,3),activation='relu', input_shape=(28,28,1)))
model.add(Conv2D(128,(3,3),activation='relu'))
model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D(2))
model.add(Dropout(0.25))

model.add(Conv2D(256,(3,3),activation='relu'))
model.add(Conv2D(128,(3,3),activation='relu'))
model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D(2))
model.add(Dropout(0.25))

model.add(Flatten())

model.add(Dense(128,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10,activation='softmax'))

model.compile(loss=tf.keras.losses.sparse_categorical_crossentropy, optimizer='adam', metrics=['accuracy'])
```

```python
earlyStop = tf.keras.callbacks.EarlyStopping(patience=10)

model.fit( x_train.to_numpy().reshape(-1,28,28,1), y_train, epochs=500,
          validation_data=( x_test.to_numpy().reshape(-1,28,28,1), y_test),
          callbacks=earlyStop
         )
```

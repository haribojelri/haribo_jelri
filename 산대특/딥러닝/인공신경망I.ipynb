```python
import tensorflow as tf
tf.__version__
```

```python
# y = x+1  
import numpy as np
import pandas as pd
```

```python
x = np.array([-3, 31, -11 ,4,0,22,-2,-5,-25,-14])
y = x+1
```

```python
x_train =  x.reshape(-1,1)
x_train
```

```python
# 신경망 설계
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
# 1. 시퀀스 생성
model = Sequential()
```

```python
# 2.층을 생성
model.add(Dense(units=1,activation='linear', input_dim = 1 ))
```

```python
# 모델의 아키텍처를  확인한다.
model.summary() 
```

```python
# 3. 모델 컴파일
# 모델을 훈련하는데 기본설정을 적용
# 손실함수(mse) 와 옵티마이져(Adam)를 적용
# metrics 옵션 : 보조평가 지표 - 손실함수를 모니터링할때 함께 추척
```

```python
# mae : 평균절대오차
# accuracy : 정확률
model.compile(optimizer='Adam',loss='mse',metrics=['mae'])
```

```python
# 4. 학습
model.fit(x_train,y,epochs=3000,verbose=1)
```

```python
# 가중치
model.weights
```

```python
from bs4 import BeautifulSoup as bs
from pprint import pprint
import requests
```

```python
html = requests.get('https://search.naver.com/search.naver?query=%EB%82%A0%EC%94%A8')
pprint(html.text)
```

```python
soup = bs(html.text,'html.parser')
soup
```

```python
# data = soup.find('div', {'class':'weekly_forecast_area _toggle_panel'})
# pprint(data)
data2 = soup.findAll('li', {'class':'week_item'})

weather_lists = [data2[i].text for i in range(10)]
weather_lists
# DataFrame 형태로. 
```

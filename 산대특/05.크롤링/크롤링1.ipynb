```python
url = 'http://www.kyobobook.co.kr/bestSellerNew/bestseller.laf?mallGb=KOR&linkClass=e&range=1&kind=0&orderClick=DAb'
html = requests.get(url)
soup = BeautifulSoup(html.content.decode('euc-kr','replace'), "html.parser")
```

```python
import re
import pandas as pd
```

```python
# findData = soup.find('ul', {'class' : 'list_type01'})
# findData.findAll('li')
findDatas = soup.select('ul.list_type01 > li')
imglist = []
titlelist = []
authorlist = []
publisher = []
publishdate = []
for data in findDatas:
    imglist.append(data.select('div.cover > a > img')[0].attrs['src']) # 이미지
    titlelist.append(data.select('div.title')[0].text.strip()) # 제목
    temp = data.select('div.author')[0].text.strip()
    temp = re.sub('\n',"",temp)
    temp = re.sub('\r',"",temp)
    temp = re.sub('\t',"",temp)
    authorlist.append(temp.split("|")[0]) # 저자    
    publisher.append(temp.split("|")[1]) # 출판사
    publishdate.append(temp.split("|")[2]) # 출판년도
df = pd.DataFrame(data=[imglist,titlelist,authorlist,publisher,publishdate]).T    
df.columns = ['link','title','author','publish','date']    
df.head()
# findDatas[0].select('div.cover > a > img')[0].attrs['src'] # 이미지
# findDatas[0].select('div.title')[0].text.strip() # 제목
# temp = findDatas[0].select('div.author')[0].text.strip()
# temp = re.sub('\n',"",temp)
# temp = re.sub('\r',"",temp)
# temp = re.sub('\t',"",temp)
# temp.split("|")[0] # 저자
# temp.split("|")[1] # 출판사
# temp.split("|")[2] # 출판년도
```
